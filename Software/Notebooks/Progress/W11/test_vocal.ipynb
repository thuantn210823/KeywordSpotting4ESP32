```python
import torch
import torchaudio
```

```python
wav1, sr = torchaudio.load('/content/00f0204f_nohash_0.wav')
wav2, sr = torchaudio.load('/content/0b09edd3_nohash_0.wav')
```

```python
f_mel = torchaudio.transforms.MelSpectrogram(sr,
                                             n_fft = 512,
                                             win_length = 480,
                                             hop_length = 160,
                                             n_mels = 40)
```

```python
mel1 = f_mel(wav1)
mel2 = f_mel(wav2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(torch.log(mel1.squeeze()+1e-6), origin = 'lower')
plt.title('Woman Voice');
```

```python
plt.imshow(torch.log(mel2.squeeze()+1e-6), origin = 'lower')
plt.title('Man Voice');
```
