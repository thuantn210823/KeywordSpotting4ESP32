```python
def add_to_class(Class):
    def wrapper(obj):
        setattr(Class, obj.__name__, obj)
    return wrapper
```

```python
class CNNet(L.LightningModule):
    def __init__(self, lr = 0.1, num_classes = 12):
        super().__init__()
        self.lr = lr
        self.net = nn.Sequential(
            nn.Conv2d(1, 3, 3, 2, 1),
            torchvision.models.mobilenet_v2(pretrained = True).features[:7],
            nn.AdaptiveAvgPool2d((1, 1)), nn.Flatten(),
            nn.Dropout(0.1),
            nn.LazyLinear(num_classes)
        )
        # self.net.apply(self.init_weights)

    def forward(self, X):
        return self.net(X)
```

```python
@add_to_class(CNNet)
def accuracy(self, Y_hat, Y, averaged = True):
    """
    Compute the number of correct predictions
    """
    Y_hat = Y_hat.reshape((-1, Y_hat.shape[-1]))
    preds = Y_hat.argmax(dim = 1).type(Y.dtype)
    compare = (preds == Y.reshape(-1)).type(torch.float32)
    return compare.mean() if averaged else compare

@add_to_class(CNNet)
def training_step(self, batch, batch_idx):
    x, y = batch
    y_hat = self.forward(x)
    loss = self.loss(y_hat, y)
    acc = self.accuracy(y_hat, y)
    sch = self.lr_schedulers()
    values = {"train_loss": loss, "train_acc": acc}
    self.log_dict(values, prog_bar = True)
    return loss

@add_to_class(CNNet)
def validation_step(self, batch, batch_idx):
    x, y = batch
    y_hat = self.forward(x)
    loss = self.loss(y_hat, y)
    acc = self.accuracy(y_hat, y)
    values = {"val_loss": loss, "val_acc": acc}
    self.log_dict(values, prog_bar = True)
    return values

@add_to_class(CNNet)
def configure_optimizers(self):
    optimizer = torch.optim.Adam(self.parameters(), self.lr)
    return optimizer

@add_to_class(CNNet)
def lr_schedulers(self):
    lr_scheduler = torch.optim.lr_scheduler.StepLR(self.configure_optimizers(), step_size = 3, gamma = 0.8)
    return lr_scheduler
```

```python
@add_to_class(CNNet)
def loss(self, y_hat, y):
    return F.cross_entropy(y_hat, y, reduction = 'mean')
```

```python
from lightning.pytorch.callbacks.early_stopping import EarlyStopping
from lightning.pytorch.callbacks import ModelCheckpoint, LearningRateMonitor
early_stopping_callback = EarlyStopping(monitor = "val_acc", min_delta = 0.001, patience = 3, mode = "max")
checkpoint_callback = ModelCheckpoint(dirpath = '/content/best_model',
                                      save_top_k = 2, monitor = 'val_acc',
                                      mode = 'max',
                                      filename = 'alex-net-mnist-{epoch:02d}-{val_loss:.2f}-{val_acc:.2f}')
lr_monitor = LearningRateMonitor(logging_interval='step')
```

```python
from lightning.pytorch import seed_everything

seed_everything(42)

net = CNNet(lr = 0.005)

trainer = L.Trainer(accelerator="gpu",
                    callbacks = [early_stopping_callback, checkpoint_callback,
                                 lr_monitor],
                    enable_checkpointing=True,
                    default_root_dir = "/content",
                    max_epochs=50)
trainer.fit(net, data)
```
