```python
import numpy as np
import urllib
import tensorflow as tf

try:
  from tensorflow.lite.experimental.microfrontend.python.ops import audio_microfrontend_op as frontend_op  # pylint:disable=g-import-not-at-top
except ImportError:
  frontend_op = None
```

```python
import torchaudio
from torchaudio.utils import download_asset

import matplotlib.pyplot as plt

import IPython.display as ipd
```

```python
SAMPLE_WAV = download_asset("tutorial-assets/Lab41-SRI-VOiCES-src-sp0307-ch127535-sg0042.wav")
```

## Read Audio

```python
SAMPLE_WAV
```

```python
wav_loader = tf.io.read_file(SAMPLE_WAV)
wav_decoder, sr = tf.audio.decode_wav(wav_loader,
                                  desired_channels = 1)
wav = wav_decoder
wav
```

```python
micro_frontend = frontend_op.audio_microfrontend(tf.cast(tf.multiply(wav, 32768), tf.int16),
                                                 sample_rate = sr,
                                                 window_size = 30,
                                                 window_step = 10,
                                                 num_channels = 40,
                                                 out_type = tf.float32)
micro_frontend.shape
```
